```python
# При реализации класса запрещено пользоваться ИИ-инструментами.

class LogisticRegression(BaseEstimator):
    """Модель логистической регрессии.

    Параметры:
    method (Literal['gd', 'sgd']): Метод оптимизации ('gd' - градиентный спуск, 
        'sgd' - стохастический градиентный спуск).
    learning_rate (float): Константа скорости обучения, на которую домножаем градиент при обучении
    tol (float): Допустимое изменение функционала между итерациями.
    max_iter (int): Максимальное число итераций.
    batch_size (int): Размер выборки для оценки градиента (используется только при 'sgd').
    fit_intercept (bool): Добавлять ли константу в признаки.
    save_history (bool): Сохранять ли историю обучения.
    """

    def __init__(
        self,
        method: Literal["gd", "sgd"] = "gd",
        learning_rate: float = 0.5,
        tol: float = 1e-3,
        max_iter: int = int(1e4),
        batch_size: int = 64,
        fit_intercept: bool = True,
        save_history: bool = True,
    ):
        """Создает модель и инициализирует параметры."""
        self.method = method
        self.learning_rate = learning_rate
        self.tol = tol
        self.max_iter = max_iter
        self.batch_size = batch_size
        self.fit_intercept = fit_intercept
        self.save_history = save_history
        self.history = []  # История обучения

    @staticmethod
    def _sigmoid(x: np.ndarray) -> np.ndarray:
        """Вычисляет сигмоидную функцию."""
        sigmoid = np.zeros(len(x))  
        idx = x >= 0
        sigmoid[idx] = 1 / (1 + np.exp(-x[idx]))
        sigmoid[~idx] = np.exp(x[~idx]) / (1 + np.exp(x[~idx]))
        return sigmoid
    def _add_intercept(self, X: np.ndarray) -> np.ndarray:
        """Добавляет свободный коэффициент к матрице признаков.

        Параметры: X (np.ndarray): Исходная матрица признаков.

        Возвращает: np.ndarray: Матрица X с добавленным свободным
        коэффициентом.
        """
        X_copy = np.full((X.shape[0], X.shape[1] + 1), fill_value=1)
        X_copy[:, :-1] = X
        return X_copy

    def logsig(self, x: np.ndarray):
        out = np.zeros_like(x)
        idx0 = x < -33
        out[idx0] = x[idx0]
        idx1 = (x >= -33) & (x < -18)
        out[idx1] = x[idx1] - np.exp(x[idx1])
        idx2 = (x >= -18) & (x < 37)
        out[idx2] = -np.log1p(np.exp(-x[idx2]))
        idx3 = x >= 37
        out[idx3] = -np.exp(-x[idx3])
        return out

    def expit_b(self, x: np.ndarray, b: np.ndarray):
        """Вычисляет sigmoid(x) - b."""
        idx = x < 0
        out = np.zeros_like(x)
        exp_x = np.exp(x[idx])
        b_idx = b[idx]
        out[idx] = ((1 - b_idx) * exp_x - b_idx) / (1 + exp_x)
        exp_nx = np.exp(-x[~idx])
        b_nidx = b[~idx]
        out[~idx] = ((1 - b_nidx) - b_nidx * exp_nx) / (1 + exp_nx)
        return out

    def grad(self, x: np.ndarray, A: np.ndarray, b: np.ndarray):
        """Вычисляет градиент функции потерь логист. регрессии.
        
        Параметры
        ----------
        x: array-like, shape (n_features,)
            Коэффициенты
    
        A: array-like, shape (n_samples, n_features)
            Матрица признаков
    
        b: array-like, shape (n_samples,)
            Метки классов
    
        Возвращает
        -------
        grad: array-like, shape (n_features,)    
        """
        z = A.dot(x)
        s = self.expit_b(z, b)
        return A.T.dot(s) / A.shape[0]
        
    def functional(self, X: np.ndarray, Y: np.ndarray, theta: np.ndarray): 
        """Вычисляет значение функции потерь логист. регрессии.
        
        Параметры
        ----------
        theta: array-like, shape (n_features,)
            Коэффициенты
    
        X: array-like, shape (n_samples, n_features)
            Матрица признаков
    
        Y: array-like, shape (n_samples,)
            Метки классов
            
    
        Возвращает
        -------
        loss: float  
            Значение функции потерь     
        """
        z = np.dot(X, theta)
        Y = np.asarray(Y)
        loss = np.mean((1-Y) * z - self.logsig(z))
        return loss
        
    def fit(self, X: np.ndarray, Y: np.ndarray) -> "LogisticRegression":
        """Обучает модель логистической регрессии.

        Также, в случае self.save_history=True, добавляет в self.history
        текущее значение оптимизируемого функционала и затраченное время.

        Параметры:
        X (np.ndarray): Матрица признаков.
        Y (np.ndarray): Вектор истинных меток.

        Возвращает:
        LogisticRegression: Обученная модель.
        """
        if X.shape[0] != Y.shape[0]:
            raise ValueError("Количество строк в X и Y должно совпадать")

        if self.fit_intercept:
            X_copy = self._add_intercept(X)
        else:
            X_copy = X.copy()

        # Заводим необходимые переменные, в которых будем сохранять инфоомацию
        iteration_count = 1
        theta = np.zeros(X_copy.shape[1])
        prev_func = 0
        curr_func = self.functional(X_copy, Y, theta)
        start_time = time.time()
        
        for i in range(self.max_iter):
            if self.method == 'gd':
                # Итерация шага стандартного градиентного спуска
                theta -= self.learning_rate * self.grad(theta, X_copy, Y)
            else:
                # Итерация шага стохастического градиентного спуска
                indices = np.random.randint(low=0, high=X_copy.shape[0], size=self.batch_size)
                X_tmp = X_copy[indices]
                Y_tmp = Y.iloc[indices]
                theta -= self.learning_rate * self.grad(theta, X_tmp, Y_tmp)

            # Фиксируем, сколько времени было потрачено на градиентный спуск
            cur_time = time.time() - start_time

            # Фиксируем значение функционала
            prev_func = curr_func
            curr_func = self.functional(X_copy, Y, theta)

            # Сохраняем в историю
            if self.save_history == True:
                self.history.append([cur_time, curr_func])
            
            iteration_count += 1
            # Смотрим, насколько сильно обновилось значение функции
            if np.abs(curr_func - prev_func) < self.tol:
                break
            
            # Делаем новую точку отсчета для времени
            start_time = time.time()

        # По разному сохраняем коэффициенты модели в зависимости от наличия свободного члена
        if (self.fit_intercept):
            self.coef_ = theta[1:]  # Коэффициенты модели
            self.intercept_ = theta[0]
        else:
            self.coef_ = theta[1:]
            self.intercept_ = 0.0 # Коэффициенты модели
        
        self.n_iter_ = iteration_count   # Число итераций

        return self

    def predict(self, X: np.ndarray) -> np.ndarray:
        """Возвращает предсказанные классы.

        Параметры: X (np.ndarray): Матрица признаков.

        Возвращает: np.ndarray: Предсказанные классы.
        """
        if self.fit_intercept:
            X_copy = self._add_intercept(X)
        else:
            X_copy = X.copy()

        if X.shape[1] != self.coef_.shape[0]:
            raise ValueError("Число признаков в X не соответствует числу коэффициентов модели")

        # Добавляем свободный член
        coef_tmp = np.zeros(X_copy.shape[1])
        coef_tmp[0] = self.intercept_
        coef_tmp[1:] = self.coef_
        
        predictions = (self._sigmoid(X_copy @ coef_tmp) >= 0.5)
        return predictions

    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        """Возвращает вероятности классов 0 и 1.

        Параметры: X (np.ndarray): Матрица признаков.

        Возвращает: np.ndarray: Матрица вероятностей классов (n_samples,
        2).
        """
        if self.fit_intercept:
            X_copy = self._add_intercept(X)
        else:
            X_copy = X.copy()

        if X.shape[1] != self.coef_.shape[0]:
            raise ValueError("Число признаков в X не соответствует числу коэффициентов модели")
            
        # Добавляем свободный член
        coef_tmp = np.zeros(X_copy.shape[1])
        coef_tmp[0] = self.intercept_
        coef_tmp[1:] = self.coef_
        
        # Соединяем столбцы вероятности  принадлежности нулевому и первому классу.
        prob_predictions = np.column_stack([1 - self._sigmoid(X_copy @ coef_tmp), self._sigmoid(X_copy @ coef_tmp)])
        return prob_predictions
```

### Задача 6.

В этой задаче вам предлагается реализовать регрессию Хьюбера, а также применить ее к данным с выбросами. Для начала реализуйте класс по шаблону снизу. Обратите внимание, что класс `HuberRegression` &mdash; наследник класса `BaseEstimator`, это с легкостью позволит использовать наш класс в различных пайплайнах библиотеки `sklearn`.

**1.** Задача оптимизации для регрессии Хьюбера выглядит следующим образом:
$$\sum_{i=1}^n R\left(Y_i - x_i^T\theta\right) \rightarrow \min_\theta,$$
где $R(x)$ &mdash; функция потерь Хьюбера, определяемая как
$$R(x) = \frac{x^2}{2} I\left\{|x| < c\right\} + c \left(|x| - \frac{c}{2}\right)I\left\{|x| > c\right\}.$$

Выпишите формулы для градиентного и стохастического градиентного спусков. В чем польза такой функции потерь?

Формула градиентного спуска:
\begin{equation}
\theta^{(t+1)} = \theta^{(t)} - \eta \sum_{i=1}^{n} \nabla R(Y_i - x_i^T \theta)
\end{equation}

Формула стохастического градиентного спуска (та же самая, но i теперь пробегает по строчкам матрицы $X_I$, выбранной случайно):
\begin{equation}
\theta^{(t+1)} = \theta^{(t)} - \eta \nabla \sum_{i=1}^{n'} R(Y_i - x_i^T \theta)
\end{equation}
Где градиент равен:
$$\nabla R(x) = x I\{|x| < c\} + c\; \text{sign}(x) I\{|x| \geq c\}$$

**2.** Реализуем теперь класс. **При реализации класса запрещено пользоваться ИИ-инструментами.**

```python
# При реализации класса запрещено пользоваться ИИ-инструментами.


class HuberRegression(BaseEstimator):
    """Класс, реализующий линейную регрессию с функцией потерь Хьюбера."""

    def __init__(self, c: float = 1.0, fit_intercept: bool = True, max_iter: int = 1000, learning_rate : float = 0.5) -> None:
        """Инициализирует модель.

        Параметры: c (float): Константа из функции потерь Хьюбера.
        fit_intercept (bool): Добавлять ли константный признак. max_iter
        (int): Максимальное число итераций оптимизации.
        """
        self.c = c
        self.fit_intercept = fit_intercept
        self.learning_rate = learning_rate
        self.max_iter = max_iter
    def _add_intercept(self, X: np.ndarray) -> np.ndarray:
        """Добавляет свободный коэффициент к матрице признаков.

        Параметры: X (np.ndarray): Исходная матрица признаков.

        Возвращает: np.ndarray: Матрица X с добавленным свободным
        коэффициентом.
        """
        X_copy = np.full((X.shape[0], X.shape[1] + 1), fill_value=1)
        X_copy[:, :-1] = X
        return X_copy      
    def f_grad(self, X: np.ndarray, y: np.ndarray, theta: np.ndarray) -> np.ndarray:
        z = y - X.dot(theta)
        mask = np.abs(z) <= self.c
        grad = -X.T.dot(mask * z + ~mask * self.c * np.sign(z))
        return grad

    def fit(self, X: np.ndarray, y: np.ndarray) -> "HuberRegression":
        """Обучает модель.

        Параметры:
        X (np.ndarray): Матрица признаков.
        y (np.ndarray): Вектор целевой переменной.

        Возвращает:
        HuberRegression: Обученная модель.
        """
        if X.shape[0] != y.shape[0]:
            raise ValueError("Количество строк в X и y должно совпадать")

        if self.fit_intercept:
            X_copy = self._add_intercept(X)
        else:
            X_copy = X.copy()
            

        theta = np.ones(X_copy.shape[1])

        number_of_iterations = 1

        for i in range(self.max_iter):
            theta -= self.learning_rate * self.f_grad(X_copy, y, theta)

        if (self.fit_intercept):
            self.coef_ = theta[1:]  # Коэффициенты модели
            self.intercept_ = theta[0]
        else:
            self.coef_ = theta[1:]
            self.intercept_ = 0.0 # Коэффициенты модели
        
        self.n_iter_ = number_of_iterations   # Число итераций

        return self

    def predict(self, X: np.ndarray) -> np.ndarray:
        """Возвращает предсказанные классы.

        Параметры: X (np.ndarray): Матрица признаков.

        Возвращает: np.ndarray: Предсказанные классы.
        """
        if self.fit_intercept:
            X_copy = self._add_intercept(X)
        else:
            X_copy = X.copy()

        if X.shape[1] != self.coef_.shape[0]:
            raise ValueError("Число признаков в X не соответствует числу коэффициентов модели")
            
        coef_tmp = np.zeros(X_copy.shape[1])
        coef_tmp[0] = self.intercept_
        coef_tmp[1:] = self.coef_
        
        pred = X_copy @ coef_tmp
        return pred
```

### Задача 7.


Рассмотрим модель одномерной регрессии $y(x) = \theta x$, где $x \in \mathbb{R}$ &mdash; одномерный признак, $y \in \mathbb{R}$ &mdash; целевой признак, $\theta \in \mathbb{R}$ &mdash; неизвестный параметр. Имеется выборка размера $n$, полученная по правилу
	$$Y_i = \theta x_i + \varepsilon_i,\ \ \ i=1,...,n,$$
где $\varepsilon_i$ &mdash; случайная ошибка измерений.

Предложите точный алгоритм поиска оценки параметра $\theta$ методом наименьших модулей, то есть $$\sum_{i=1}^n \left|Y_i - \theta x_i\right| \to \min_\theta,$$ работающий за время $O(n \log n)$. Приведите его описание и теоретическое обоснование. Реализация в коде не требуется.

**Решение:**

1. Приведем сумму к такому виду, пользуясь тем, что все $x_i$ - ненулевые, так как иначе это слагаемое неинтересно в оптимизации $F(\theta) = \sum_{i=1}^n \left|Y_i - \theta x_i\right| = \sum_{i=1}^n|x_i|\cdot|\frac{Y_i}{x_i} - \theta| = \sum_{i=1}^n w_i|a_i-\theta|, \; w_i \geq 0$
2. Создадим массив $S = [a_1, a_2, ..., a_n]$, а затем отсортируем его, это работает за $O(n\log{n})$. Отсортированный массив назовем $S_{sorted} = [b_1, b_2, ..., b_n] = Sort(S)$
3. Также создадим массив $Weights = [w_1, w_2, ... w_n]$, где $w_i$ - вес $b_i$. И создадим массив префиксных и суффиксных сумм массива $Weights$: $Psumm$ и $Ssumm$ соответственно. Это все отработает за $O(n)$
4. Т.к $F(\theta)$ - сумма модулей, то функция $F(\theta)$ - возрастает, когда сумма коэффициентов при $\theta$ после раскрытия модулей больше нуля и убывает в другом. Тогда понятно, что нам нужно найти такое $\theta$, что сумма коэффициентов впервые становится больше нуля. Понятно, что это $\theta \in \{a_1, a_2, ... a_n\}$, т.к только при значениях меняется сумма коэффициентов при $\theta$.
5. Далее пробегаемся циклом `for i in range(n)` и считаем, когда $Psumm$ - $Ssumm$ станет больше нуля. Эта разность и будет отвечать за коэффициент при $\theta$, ведь все элементы, которые левее $b_i$ дают вклад с плюсом (в силу отсортированности), а все элементы правее левого - дают вклад с минусом. Значение $\theta$, при котором впервые сумма будет больше нуля и будет искомым. Понятно, что это все тоже работает за $O(n)$
6. Так как все пункты работают не более, чем за $O(n\log{n})$, то наш алгоритм работает за $O(n\log{n})$
